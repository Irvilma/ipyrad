#### Starting an analysis (CLI) 
The first thing we do is to create a params file for the ipyrad analysis. You can do this at the command line terminal with the (-n) argument, like below. Here we pass it the name "Rontest" for our assembly.

```python
%%bash
ipyrad -n "Rontest"
```

#### Alternate start (API)
Here I will instead use the API to create the params file since it is more programmatic 
to save the settings in this notebook using code rather than a text editor. I simply set the parameters we
want to change and then save the Assembly under the name "Rontest". 

```python
## import ipyrad
import ipyrad as ip

## create an Assembly object
data = ip.Assembly("Rontest")

## set params for Rontest
data.set_params(1, "Rontest")
data.set_params(4, "/home/deren/Downloads/Ron/*.gz")
data.set_params(7, "pairddrad")
data.set_params(8, ",")
data.set_params(25, (6, 4))

## save the params to a file, use force to overwrite existing
data.write_params(force=True)

## print the changed params dict.
data.get_params()
```
